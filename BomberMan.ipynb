```python
r=6
c=7
n=5
grid=['.......','...O.O.','....O..','..O....','OO...OO','OO.O...']

'''
 the hacker code expects ['.......','...O.O.','...OO..','..OOOO.','OOOOOOO','OOOOOOO']
 output doesn't match, but logically my answer seems correct. 
 Bombs will be planted in most slots of most rows in t=2, explode in t=5.
 ['.......', '.......', '.......', '.......', '.......', 'O......']
'''
```

1. Initially, Bomberman arbitrarily plants bombs in some of the cells, the initial state.
2. After one second, Bomberman does nothing.
3. After one more second, Bomberman plants bombs in all cells without bombs, thus filling the whole grid with bombs. No bombs detonate at this point.
4. After one more second, any bombs planted exactly three seconds ago will detonate. Here, Bomberman stands back and observes.
5. Bomberman then repeats steps 3 and 4 indefinitely.

Initialize empty dictionary 

Bombs={}

t=0, initial grid; Make a list of bombs in initial state Bombs[0]

t=1, nothing

t=2, plant new bombs in this state Bombs[2]


t=3, Bombs[0] blow

t=4, plant new bombs in this state Bombs[4]


t=5, Bombs[2] blow

t=6, plant new bombs in this state Bombs[6]


t=7, Bombs[4] blow

t=8, plant new bombs in this state Bombs[8]


t=9, Bombs[6] blow

t=10, plant new bombs in this state Bombs[10]


etc.

```python
Bombs={}
Grids={}
```

```python
# Make functions for step 3 and 4
def plant(t):
    grid_p=Grids[t-1].copy()
    Bombs[t]=[]
    for row in range(r):
        # plan bombs to be planted at t, Bombs[t]
        # if the character is '.' we will replace it with a bomb
        for col in range(c):
            if grid_p[row][col]=='.':
                Bombs[t].append([row,col]) 
        # plant bombs
        grid_p[row]=str.replace(grid_p[row],'.','O')
    #keep the memory trim by removing what we have used up
    Grids.pop(t-1)
    return grid_p
```

```python
def blow(t):
    grid_b=Grids[t-1].copy()
    for bomb in Bombs[t-3]:
        row=bomb[0]
        col=bomb[1]
        # for each bomb location make 4 adjacent cells '.'
        for i in range(r):
            if i==row:
                new_row=''
                for j in range(c):
                    if (j==col or j==(col-1) or j==(col+1)):
                        new_row=new_row+'.'
                    else:
                        new_row=new_row+grid_b[i][j]
                grid_b[i]=new_row
            elif (i==(row-1) or i==(row+1)):
                new_row=''
                for j in range(c):
                    if (j==col):
                        new_row=new_row+'.'
                    else:
                        new_row=new_row+grid_b[i][j]
                grid_b[i]=new_row
    #keep the memory trim by removing what we have used up
    Bombs.pop(t-3)
    Grids.pop(t-1)
    return grid_b
```

```python
Grids[0]=grid.copy()
Bombs[0]=[]

for row in range(r):
    for col in range(c):
        if grid[row][col]=='O':
            Bombs[0].append([row,col])
Grids[1]=Grids[0].copy()

print('Grids[ 0 ]:',Grids[0])
print('Grids[ 1 ]:',Grids[1])

for t in range(2,n+1,2):
    Grids[t]=plant(t)
    Grids[t+1]=blow(t+1)
```
