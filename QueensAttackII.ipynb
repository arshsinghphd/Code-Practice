```python
''' Input
5 3
4 3
5 5
4 2
2 3
Expect 10
'''

'''
4 0
4 4
expect 9
'''

'''
1 0
1 1
expect 0
'''
```

```python
n=5 # grid size
k=3 # no. of obstacles
r_q=4 # row loc. queen
c_q=3 # col. loc. queen
obstacles=[[5,5],[4,2],[2,3]] # obstacles locations
# expect 10 as answer
```

```python
# Define all possible moves as functions move1-8
# if the move results in moving outside the grid
    #the functions return the same box that was sent. 
def move1(a):
    if ((a[1]+1)<=n):
        return [a[0],a[1]+1]
    else:
        return a
def move2(a):
    if (((a[0]+1)<=n) and ((a[1]+1)<=n)):
        return [a[0]+1,a[1]+1]
    else:
        return a
def move3(a):
    if ((a[0]+1)<=n):
        return [a[0]+1,a[1]]
    else:
        return a
def move4(a):
    if (((a[0]+1)<=n) and ((a[1]-1)>=1)):
        return [a[0]+1,a[1]-1]
    else:
        return a
def move5(a):
    if ((a[1]-1)>=1):
        return [a[0],a[1]-1]
    else:
        return a
def move6(a):
    if (((a[0]-1)>=1) and ((a[1]-1)>=1)):
        return [a[0]-1,a[1]-1]
    else: 
        return a
def move7(a):
    if ((a[0]-1)>=1):
        return [a[0]-1,a[1]]
    else:
        return a
def move8(a): 
    if (((a[0]-1)>=1) and ((a[1]+1)<=n)):
        return [a[0]-1,a[1]+1]
    else:
        return a
```

```python
# attack is a dictionary with all move functions 1-8
attack={1:move1,2:move2,3:move3,4:move4,5:move5,6:move6,7:move7,8:move8}

# position a of queen as a tuple
a=(r_q,c_q)

# count the boxes where queen can move
moves=[]
counter=0

# Check all directions to see if queen can move one box. 
    # If the function returns the same box,
    # this is our signal to not move in that direction further.
    # i.e. pop the move in that direction from dictionary attack.
for move in attack:
    while(attack[move](a)!=a):
        b=attack[move](a)
        if (b not in obstacles) and (b not in moves):
            moves.append(b)
            counter+=1
            a=b
        else: 
            break
    a=(r_q,c_q)
print(counter)
```
