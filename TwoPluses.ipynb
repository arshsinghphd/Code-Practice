```python
import random
random.seed(56)
n=random.randint(2,15)
m=n
grid=[]
for i in range(n):
    string=''
    for j in range(m):
        for char in random.sample(['G','G'],1):
            string+=char
    grid.append(string)
```

```python
for _ in grid:
    print(_)
```

```python
def getGrid(i,j,temp):
    if (0<=i<n) and (0<=j<m):
        return temp[i][j]
    else:
        return 'B'
```

```python
def test(i,j,nn,temp):
    result=True
    for c in range(1,nn+1):
        if (getGrid(i+c,j,temp)=='G' and 
            getGrid(i-c,j,temp)=='G' and
            getGrid(i,j+c,temp)=='G' and
            getGrid(i,j-c,temp)=='G'):
            result=True
        else:
            result=False
            break
    return result
```

```python
def turnB(i,j,nn,temp):
    bs=str('B'*(2*nn+1))
    temp[i]=temp[i][:j-nn]+bs+temp[i][j+nn+1:]
    for c in range(1,nn+1):
        temp[i+c]=temp[i+c][:j]+'B'+temp[i][j+1:]
        temp[i-c]=temp[i+c][:j]+'B'+temp[i][j+1:]
    return temp
```

```python
def twoPluses(n,m,grid):
    ma=round(min(n/2,m/2))
    ans=[]
    for max_nn in range(ma,0,-1): # This loop to find max ans
        temp=grid.copy()
        pluses=[]
        #print('max_nn:',max_nn)
        for nn in range(max_nn,-1,-1):
            for i in range(n):
                for j in range(m):
                    if temp[i][j]=='G' and len(pluses)<2:
                            t=test(i,j,nn,temp)
                            if t:
                                pluses.append(4*nn+1)
                                #print(i,j,nn,pluses,temp)
                                temp=turnB(i,j,nn,temp)
        if len(pluses)>1:
            ans.append(pluses[0]*pluses[1])
        
        #print(ans)
    return max(ans)
```

```python
mans=twoPluses(n,m,grid)
print(mans)
```
